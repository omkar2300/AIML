```python
import numpy as np

# Create array
arr = np.array([1, 2, 3, 4, 5])

print("Array:",arr)

#operations
print("mutiply by 2:", arr*2)

#statistics
print("mean:", np.mean(arr))
print("sum:", np.sum(arr))
```

```python
import pandas as pd

# Create dataset

data = {
    "name" :["omkar", "sachin", "rahul"],
    "age" : [25, 30, 28],
    "salary" : [50000, 60000, 55000]
}

df = pd.DataFrame(data)

print(df)

#basic operations
print("average salary:", df["salary"].mean())
print("max age:", df["age"].max())
```

```python
import pandas as pd

# Create Dataset

data = { 
    "Student_name" : ["omkar", "sachin", "rahul"],
    "marks" : [45, 50, 48],

    "products" : ["laptop", "mobile", "tablet"],
    "prices" : [50000, 20000, 15000]

}

df = pd.DataFrame(data)

print(df)

print("average marks:", df["marks"].mean())
print("min marks:", df["marks"].min())

print("min price:", df["prices"].min())
print("average price:", df["prices"].mean())



```
